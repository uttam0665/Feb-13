Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

ans:-
    The ArithmeticError class in Python is a built-in exception class that is raised when an arithmetic operation encounters an error.
    This class is a subclass of the Exception class and serves as the base class for several specific arithmetic exception classes.
    Some of the errors defined in the ArithmeticError class are:

        ZeroDivisionError: This error is raised when attempting to divide a number by zero.


OverflowError: This error is raised when a calculation exceeds the maximum limit for a numeric type.
example:-
    



```python
#  OverflowError

try:
    result = 10**1000
except OverflowError as e:
    print("Error:", e)

# ZeroDivisionError

try:
    result = 10 / 0
except ZeroDivisionError as e:
    print("Error:", e)


```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
ans:-

The LookupError class in Python is used as a base class for exceptions that occur during lookup operations, where a key or index is not found. It provides a common base for more specific lookup-related exception classes. Let's explore two commonly used exceptions derived from LookupError: KeyError and IndexError.
 

```python
#KeyError: This exception is raised when a dictionary key is not found.

#example:

my_dict = {'name': 'John', 'age': 30}

try:
    value = my_dict['gender']
except KeyError as e:
    print("Error:", e)


#IndexError: This exception is raised when a sequence index is out of range.

# example

my_list = [1, 2, 3]

try:
    value = my_list[5]
except IndexError as e:
    print("Error:", e)





```

Q6. List down some best practices for exception handling in python.
    ans:-
        
      
Certainly! Here are some best practices for exception handling in Python:

Be specific in exception handling: Catch specific exceptions rather than using a generic except block. This allows you to handle different exceptions differently and provides more accurate error handling.

Use multiple except blocks: When handling multiple exceptions, use separate except blocks for each exception. This allows you to handle each exception individually and provide specific error messages or actions.

Avoid bare except statements: Avoid using bare except statements without specifying the exception type. This can make it difficult to identify and debug specific errors and may lead to unexpected behavior.

Use finally for cleanup: Use the finally block to ensure that cleanup or resource release operations are always executed, regardless of whether an exception occurred or not.

Keep exception handling minimal: Only catch exceptions that you can handle effectively. Avoid catching exceptions that you can't handle properly or that should propagate to a higher level of the program for appropriate handling.

Log exceptions: Use a logging framework to log exceptions and related information. This helps in debugging and troubleshooting issues in production environments.

Reraise exceptions selectively: If you catch an exception but cannot handle it appropriately, consider reraising the exception using the raise statement. This allows the exception to propagate to a higher level where it can be handled more effectively.

Use custom exceptions when necessary: Create custom exceptions for specific error conditions in your application. This provides clarity, modularity, and allows for more granular error handling.

Follow the EAFP principle: Pythonic code often follows the "Easier to Ask for Forgiveness than Permission" (EAFP) principle. Instead of checking for conditions before performing an action, try performing the action and catch and handle any resulting exceptions.

Document exception handling: Document your exception handling strategy and any specific exceptions that your code might raise. This helps other developers understand how to handle exceptions and provides guidance for maintaining and extending your code.
